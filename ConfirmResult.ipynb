```python
import numpy as np
import matplotlib.pyplot as plt

folderpath = "results/informer_ETTh1_ftM_sl96_ll48_pl24_dm512_nh8_el2_dl1_df2048_atprob_fc5_ebtimeF_dtTrue_mxTrue_test_0"

pred =\
np.load(folderpath+"/real_prediction.npy")

true =\
np.load(folderpath+"/real_prediction_trues.npy")

pred0 = pred[0]
pred7 = [pred0[i][6] for i in range(24)]

true0 = true[0]
true7 = [true0[i][6] for i in range(24)]

plt.plot(pred7,linestyle = "dotted",label = "prediction")
plt.plot(true7,label = "true")
plt.legend()
```
