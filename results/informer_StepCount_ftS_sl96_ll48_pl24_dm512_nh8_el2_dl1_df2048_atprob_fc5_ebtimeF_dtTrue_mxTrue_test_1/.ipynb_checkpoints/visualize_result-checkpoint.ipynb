```python
import numpy as np
import matplotlib.pyplot as plt

pred = np.load("real_prediction.npy")

true = np.load("real_prediction_trues.npy")
pred.shape
```

```python
pred0 = pred[0]
true0 = true[0]
pred0.shape
```

```python
pred_reshape = pred0.reshape([24])
true_reshape = true0.reshape([24])

pred_reshape.shape
```

```python
metrics = np.load("metrics.npy")
metrics
mae, mse, rmse, mape, mspe = metrics
```

```python
condition = "informer_StepCount_ftS_sl96_ll48_pl24_dm512_nh8\n_el2_dl1_df2048_atprob_fc5_ebtimeF_dtTrue_mxTrue_test_1\n"

metrincs = "mae_"+str(mae)+ "_mse_"+str(mse)

plt.plot(pred_reshape,linestyle = "dotted",label = "prediction")
plt.plot(true_reshape,label = "true")
plt.legend()
plt.title(condition+metrincs)
plt.savefig(condition+metrincs+".png", dpi=300)
```
