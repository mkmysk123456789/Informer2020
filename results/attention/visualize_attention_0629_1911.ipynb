```python
import numpy as np
import matplotlib.pyplot as plt


attention = np.load('attention_weight.npy')

attention = attention[0]
attention  = np.sum(attention,axis=0)

attention = attention[48:,:]

attention = attention.T

min = attention.min(axis=1, keepdims=True)
max = attention.max(axis=1, keepdims=True)
attention = (attention-min)/(max-min)


target = np.load('attention_target.npy')
target = target[0,:,0]


min = target.min(axis=0, keepdims=True)
max = target.max(axis=0, keepdims=True)
target = (target-min)/(max-min)


print(attention.shape)
print(target.shape)


plt.subplots(figsize=(20.0, 10.0))

for i in range(10):
    plt.plot(attention[:,-i],linestyle = "dotted",label = "attention_weight_-"+str(i))

plt.plot(target,label = "encorder_input_raw_stepcount")
plt.legend()
plt.savefig("visualize_attention_0629_1911.png", dpi=300)
```
