```python
import numpy as np

attention = np.load('attention_weight.npy')
```

```python
attention.shape
```

```python
attention = attention[0]
attention.shape
```

```python
attention  = np.sum(attention,axis=0)
attention.shape
```

```python
attention  = np.sum(attention,axis=1)
attention.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(attention)
```

```python
target = np.load('attention_target.npy')
target.shape
target = target[0,:,0]

# plt.plot(attention,linestyle = "dotted",label = "attention_weight")
plt.plot(target,label = "target_value")
plt.legend()
# plt.savefig".png", dpi=300)
```

```python
min = attention.min(axis=0, keepdims=True)
max = attention.max(axis=0, keepdims=True)
attention = (attention-min)/(max-min)
plt.plot(attention)
```

```python
min = target.min(axis=0, keepdims=True)
max = target.max(axis=0, keepdims=True)
target = (target-min)/(max-min)
plt.plot(target)
```

```python
plt.plot(attention,linestyle = "dotted",label = "attention_weight")
plt.plot(target,label = "target_value")
plt.legend()
```
