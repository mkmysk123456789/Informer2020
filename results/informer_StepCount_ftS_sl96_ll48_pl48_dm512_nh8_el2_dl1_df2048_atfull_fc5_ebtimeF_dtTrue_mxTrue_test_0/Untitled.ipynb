```python
import numpy as np
import matplotlib.pyplot as plt

pred_len = 48

pred = np.load("pred_preds.npy")

true = np.load("pred_trues.npy")

pred_reshape = pred[0].reshape([pred_len])
true_reshape = true[0].reshape([pred_len])


metrics = np.load("metrics.npy")
metrics
mae, mse, rmse, mape, mspe = metrics


condition = "informer_StepCount_ftS_sl96_ll48_pl48_dm512_nh8_el2_dl1_df2048_atprob_fc5_ebtimeF_dtTrue_mxTrue_test_1\n"

metrincs = "mae_"+str(mae)+ "_mse_"+str(mse)

plt.plot(pred_reshape,linestyle = "dotted",label = "prediction")
plt.plot(true_reshape,label = "true")
plt.legend()
plt.title(condition+metrincs)
plt.savefig(condition+metrincs+".png", dpi=300)
```
